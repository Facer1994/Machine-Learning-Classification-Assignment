```python
import pandas as pd
```

```python
dataset=pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset= pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes',]]
independent
```

```python
dependent= dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_= sc.fit_transform(x_train)
X_test_= sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
import lightgbm as ltb
param_grid={'boosting_type':['gbdt','dart'],'n_estimators':[100,200,300],'max_depth':[15,30,45,60],'num_leaves':[10,20,30,40],'learning_rate':[0.1],'random_state':[None]}
grid= GridSearchCV(ltb.LGBMClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1)   
grid.fit(independent,dependent) 
```

```python
re=grid.cv_results_
re
print(re)
grid_predictions = grid.predict(X_test_) 
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)



# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```
